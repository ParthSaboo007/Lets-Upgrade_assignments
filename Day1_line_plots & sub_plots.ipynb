```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.plot(a,b, linewidth=5,markersize=8)
plt.title('Line plot', size=25)
plt.xlabel('x axis', size=20)
plt.ylabel('y axis', size=20)
```

```python
# Sales trend in a company line plot
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

plt.plot(sales_1,days,linewidth=4, markersize=5, label='sales of company 1')
plt.plot(sales_2,days,linewidth=5,label="sales of company 2")
plt.legend()
plt.title('Line plots',size=20)
plt.xlabel('sales',size=20)
plt.ylabel('days of week', size=20)
```

```python
# plotting subplots
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplot(3,3,1)
plt.plot(x,y1)
plt.subplot(3,3,2)
plt.plot(x,y2)
plt.subplot(3,3,3)
plt.plot(x,y3)
plt.subplot(3,3,4)
plt.plot(x,y4)
plt.subplot(3,3,5)
plt.plot(x,y5)
```
